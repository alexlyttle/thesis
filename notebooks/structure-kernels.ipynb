```python
model = gong.merge(pulse, combine_attrs="drop_conflicts")
model
```

```python
K = structure_kernels(model)
```

```python
def plot_kernel(model, K, l=0, n_pg=21, inverse=False, ax=None):
    if ax is None:
        _, ax = plt.subplots(figsize=(6, 4))
    
    i = 1 if inverse else 0
    y = model.R*K[i].sel(n_pg=n_pg, l=l)
    ax.plot(model.tau, y)
    return ax
```

```python
conv_zone = model.where(conv)

fig, ax = plt.subplots()
for n in [5, 15, 25]:
    plot_kernel(conv_zone, K["c2_rho"], n_pg=n, ax=ax)
```

```python
fig, ax = plt.subplots()
for n in [5, 15, 25]:
    plot_kernel(conv_zone, K["c2_rho"], n_pg=n, inverse=True, ax=ax)
```

```python
fig, ax = plt.subplots()
for n in [5, 15, 25]:
    plot_kernel(conv_zone, K["Y_u"], n_pg=n, ax=ax)
```

```python
from thesis.calculus import integrate
```

```python
def adiabatic_change(tau, tau_HeII, Delta_HeII, Gamma_HeII):
    return (
        - Gamma_HeII / np.sqrt(2*np.pi) / Delta_HeII 
        * np.exp(- (tau - tau_HeII)**2/2/Delta_HeII**2)
    )
```

```python
def plot_kernel(model, K, l=0, n_pg=21, inverse=False, ax=None, **kwargs):
    if ax is None:
        _, ax = plt.subplots(figsize=(6, 4))
    
    i = 1 if inverse else 0
    y = model.R*K[i].sel(n_pg=n_pg, l=l)
    ax.plot(model.tau/model.tau.max(), y, **kwargs)
    return ax
```

```python
l = 0
n = [15, 21, 26]
num_modes = len(n)

fig, axes = plt.subplots(num_modes, figsize=(6, 8), sharex=True, sharey=True)

for i, ni in enumerate(n):
    ax = axes[i]
    ax = plot_kernel(model, K["c2_rho"], l=l, n_pg=ni, ax=ax, label=r"$c^2, \rho$")
    ax = plot_kernel(model, K["c2_rho"], l=l, n_pg=ni, inverse=True, ax=ax, label=r"$\rho, c^2$")
    ax.set_ylabel("scaled structure kernel")
    if i == 0:
        ax.legend(loc="upper right")
        
    ax.set_title(f"$n = {ni}$", y=1.0, pad=-18)
    # ax = ax.twinx()
    # ax.plot(model.tau/model.tau.max(), differentiate(model.cs, model.tau), color="gray", ls="--")
    # ax.set_ylim(0)
    # ax.set_yticks([])

axes[-1].set_xlabel("fractional acoustic depth")
fig.tight_layout()
fig.savefig("../figures/kernels.pdf", format="pdf", dpi=300)
```

```python
from thesis.calculus import integrate
```

```python
def adiabatic_change(tau, tau_HeII, Delta_HeII, Gamma_HeII):
    return (
        - Gamma_HeII / np.sqrt(2*np.pi) / Delta_HeII 
        * np.exp(- (tau - tau_HeII)**2/2/Delta_HeII**2)
    )
```

```python
Gamma_HeII = 10.0
Delta_HeII = 80.0
tau_HeII = 800.0

gauss1 = adiabatic_change(model.tau, tau_HeII, Delta_HeII, Gamma_HeII)
gauss2 = adiabatic_change(model.tau, tau_HeII*0.7, Delta_HeII*0.9, Gamma_HeII*0.4)

# dnu = integrate(K["c2_rho"][0]*dG1_G1, model.r, axis=0)
dnu = integrate(K["c2_rho"][0]*dG1_G1_model, model.r, axis=0)

dnu_gauss1 = integrate(K["c2_rho"][0]*gauss1, model.r, axis=0)
dnu_gauss2 = integrate(K["c2_rho"][0]*gauss2, model.r, axis=0)
dnu_gaussT = integrate(K["c2_rho"][0]*(gauss2+gauss1), model.r, axis=0)

plt.plot(model.freq.real, dnu*model.freq.real, "oC0")
plt.plot(model.freq.real, dnu_gauss1*model.freq.real, "oC1")
# plt.plot(model.freq.real, dnu_gauss2*model.freq.real, "oC2")
plt.plot(model.freq.real, dnu_gaussT*model.freq.real, "oC2")
plt.xlim(1000, 4000)
plt.ylim(-10, -1)
```
